#### Deleting the columns not required

```python
columns_notreq=['ID','Name','Photo','Flag','Club Logo','Jersey Number','Joined','Loaned From','Contract Valid Until']
data=data.drop(data[columns_notreq],axis=1)
```

##### Transforming the object data type to float in Value, Wage, Release Clause attributes

```python


def currency_to_float(value):
    if isinstance(value, float):
        return value  # Return the float value as is
    elif 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1e3
    elif 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1e6
    else:
        return float(value.replace('€', ''))

data['Value'] = data['Value'].apply(currency_to_float)
data['Wage'] = data['Wage'].apply(currency_to_float)
data['Release Clause'] = data['Release Clause'].apply(currency_to_float)
```

```python
data.head()
```

##### Transforming Body Type,Height, Weight attributes

```python
def bodytype(value):
    if (value=='Normal'or value=='Lean' or value=='Stocky'):
        return value
    else:
        return None

data['Body Type'] = data['Body Type'].apply(bodytype)
```

```python
data['Body Type']
```

```python
def height_to_cm(height):
    if isinstance(height, float):
        return height
    else:
        feet, inches = height.split("'")
        feet_cm = int(feet) * 30.48
        inches_cm = int(inches) * 2.54
        return feet_cm + inches_cm
data['Height'] = data['Height'].apply(height_to_cm)

def weight_to_kg(weight):
    if isinstance(weight, float):
        return weight
    else:
        pounds = float(weight.replace('lbs', ''))
        kg = pounds * 0.45359237  # 1 lb ≈ 0.45359237 kg
        return kg
data['Weight'] = data['Weight'].apply(weight_to_kg)
```

##### Splitting the Work Rate attribute

##### Updating the required columns in the proper format 

```python
def transform_column(value):
    if isinstance(value, float):
        return value
    else:
        parts = value.split('+')
        if len(parts) == 2:
            base_value = int(parts[0])
            additional_value = int(parts[1])
            new_value = base_value + additional_value
            return float(new_value)
        else:
            return value
columns_to_transform = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB','LB','LCB','CB','RCB','RB']
for column in columns_to_transform:
    data[column] = data[column].apply(transform_column)
```
